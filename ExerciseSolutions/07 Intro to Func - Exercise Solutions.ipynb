## Functions

#### Greeter
- Write a function that takes in a person's name, and prints out a greeting.
    - The greeting must be at least three lines, and the person's name must be in each line.
- Use your function to greet at least three different people.
- **Bonus:** Store your three people in a list, and call your function from a `for` loop.

```python
# Ex 4.1 : Greeter

def greet(name: str):
    print(f'Good Morning, {name}!')
    print(f'How has your day been today, {name}?')
    print(f'Would you like anything, {name}?')
    
listOfNames: list = ['Ron', 'Harry', 'Hermoine']

for n in listOfNames:
    greet(n)
    print('')
```

#### Full Names
- Write a function that takes in a first name and a last name, and prints out a nicely formatted full name, in a sentence. Your sentence could be as simple as, "Hello, *full\_name*."
- Call your function three times, with a different name each time.

```python
# Ex 4.2 : Full Names

def hello(Fname: str, Lname: str):
    print(f'Hello, {Fname.title()} {Lname.title()}!')
    
listOfNames: list = [['Ron', 'Weasley'], ['Harry', 'Poter'], ['Hermoine', 'Granger']]

for names in listOfNames:
    hello(names[0], names[1])
```

#### Addition Calculator
- Write a function that takes in two numbers, and adds them together. Make your function print out a sentence showing the two numbers, and the result.
- Call your function with three different sets of numbers.

```python
# Ex 4.3 : Addition Calculator

def add(num1: int, num2: int):
    print(f'{num1} + {num2} = {num1 + num2}')
    
numbers: list = [[1, 2], [3, 4], [5, 6]]

for nums in numbers:
    add(nums[0], nums[1])
```

#### Return Calculator
- Modify *Addition Calculator* so that your function returns the sum of the two numbers. The printing should happen outside of the function.

```python
# Ex 4.4 : Return Calculator

def add(num1: int, num2: int) -> int:
    return (num1 + num2)
    
numbers: list = [[1, 2], [3, 4], [5, 6]]

for nums in numbers:
    sum = add(nums[0], nums[1])
    print(f'{nums[0]} + {nums[1]} = {sum}')
```

#### List Exercises - Functions
- Go back and solve each of the following exercises from the section on [Lists and Tuples](http://nbviewer.ipython.org/urls/raw.github.com/ehmatthes/intro_programming/master/notebooks/lists_tuples.ipynb), using functions to avoid repetetive code:
    - [Ordered Working List](http://nbviewer.ipython.org/urls/raw.github.com/ehmatthes/intro_programming/master/notebooks/lists_tuples.ipynb#exercises_common_operations)
    - [Ordered Numbers](http://nbviewer.ipython.org/urls/raw.github.com/ehmatthes/intro_programming/master/notebooks/lists_tuples.ipynb#exercises_common_operations)

```python
# Ex 4.5 : List Exercises - Functions
# Ordered Working List

def printList(lis: list):
    for i in range(len(lis)):
        print(f"\t{i+1}. {lis[i]}")


careers: list = ['athlete', 'programmer', 'driver', 'designer', 'doctor', 'artist']

print("original:")
printList(careers)
print("\nalphabetical:")
printList(sorted(careers))

print("\n\noriginal:")
printList(careers)
print("\nreverse alphabetical:")
printList(sorted(careers, reverse=True))

print("\n\noriginal")
printList(careers)
careers.reverse()
print("\nreverse:")
printList(careers)
careers.reverse()

print("\n\noriginal:")
printList(careers)
careers.sort()
print("\npermanent alphabetical:")
printList(careers)
careers.sort(reverse=True)
print("\npermanent reverse alphabetical:")
printList(careers)
```

```python
# Ex 4.5 : List Exercises - Functions
# Ordered Numbers

def printList(lis: list):
    for i in range(len(lis)):
        print(f"\t- {lis[i]}")
        

numbers: list = [12, 34, 21, 54, 23]
    
print("original:")
printList(numbers)
print("\nincreasing order:")
printList(sorted(numbers))

print("\n\noriginal:")
printList(numbers)
print("\ndecreasing order:")
printList(sorted(numbers, reverse=True))

print("\n\noriginal:")
printList(numbers)
numbers.reverse()
print("\nreverse order:")
printList(numbers)
numbers.reverse()

print("\n\noriginal:")
printList(numbers)
numbers.sort()
print("\npermanent increasing order:")
printList(numbers)
numbers.sort(reverse=True)
print("\npermanent decreasing order:")
printList(numbers)
```

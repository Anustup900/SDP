##### People From different Places

```python
for i in ['S','Q','C']:
    city = None
    if i == 'S':
        city = "Southampton"
    elif i == 'Q':
        city = "Cherbourg"
    elif i== 'C':
        city = "Queenstown"
    
    count = cTitanic[cTitanic['Embarked']==i]
    print("People Count From {} is {}".format(city,count.shape[0]))
    survived = count[count["Survived"]==1]
    print("People from {} survived:{}".format(city,survived.shape[0]))
    print("-----------------------------------------------------------")
    print("-----------------------------------------------------------")
```

### Exploration of Data on the Basis of Parch
<p id="Parch">Below is the given Exploration</p>

```python
cTitanic.head()
```

```python
# checking for distinct 
def check(li,a):
    if(len(li)==0):
        return False
    for i in li:
        if i==a:
            return True
        
    return False

dist = []
insert = False
for i in cTitanic["Parch"]:
    insert = check(dist,i)
    if(not(insert)):
        dist.append(i)

print(dist)
```

```python
dist = sorted(dist)
for i in dist:
    co = cTitanic[cTitanic["Parch"]==i]
    print("Number of People with %d Children %d"%(i,co.shape[0]))
```

<h2><p id="Pclass">Exploring Data on the Basis of Pclass</p></h2>

```python
pl = [1,2,3]
for i in pl:
    data = cTitanic[cTitanic['Pclass']==i]
    print("People from class {} are {}".format(i,data.shape[0]))
    print("People from class {} Survived are {}".format(i,data[data.Survived==1].shape[0]))
    print("Percentage Survived From Class",i,"are -:",(data[data.Survived==1].shape[0]/data.shape[0])*100)
    print("-------------------------------------------------------------------------------------------------------------")
```
